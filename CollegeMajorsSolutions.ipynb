```python
#import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_WomenSTEM = pd.read_csv("women-stem.csv") #load in csv file

df_WomenSTEM.head() #displays first five lines; can specify number of lines inside parenthese (ex: df.head(10) shows first 10 lines)
```

```python
#get general info about data: names of columns, number of entries per column, data type (int, string, etc)
df_WomenSTEM.info()
```

```python
#counts of unique values in "Major_category" column, including NaN values
df_WomenSTEM["Major_category"].value_counts(dropna=False)
```

```python
#sum of null values per column; in this case the data set is clean, we don't need to do anything
pd.isnull(df_WomenSTEM).sum()
```

```python
#find value type for each column
df_WomenSTEM.dtypes 
```

```python
df_WomenSTEM.groupby(["Major_category"])[["Median"]].mean() 
```

```python
major_women_gb = df_WomenSTEM.groupby(["Major_category"])[["ShareWomen"]].mean()
major_women_gb #created a variable so I can use later while graphing
```

```python
df_WomenSTEM.corr() #table of correlation between all columns
```

```python
#graphing as bar graph again, using pandas

major_women_gb.plot.bar(title="Percentage of Women in STEM Majors", color="navy")
```

```python
#plotting as scatter plot, using matplotlib
sharewomen = df_WomenSTEM["ShareWomen"] #saving ShareWomen series as a variable, to use later
womenmedian = df_WomenSTEM["Median"] #new variable for another series, Median (salary)

plt.scatter(list([sharewomen]), list([womenmedian]))#converted series into lists directly, and plotted the lists
plt.show()
```
